## 3. CCTV 데이터 정리하기

### 3.1 데이터의 컬럼 이름을 변경해보자

```python
#데이터의 열이름 변경:
#'자치구' -> '구별',
#'합계' -> '발생합계', '합계.1' -> '검거합계',
#'살인.1'-> '살인검거',
#'강도.1' -> '강도검거',
#'강간강제추행.1' -> '강간강제추행검거',
#'절도.1' -> '절도검거',
#'폭력.1' -> '폭력검거'

# 데이터프레임의 열 이름 변경
Crime_Seoul.rename(columns={
    '자치구': '구별',       # '자치구'를 '구별'로 변경
    '합계': '발생합계',     # '합계'를 '발생합계'로 변경
    '합계.1': '검거합계',   # '합계.1'을 '검거합계'로 변경
    '살인.1': '살인검거',   # '살인.1'을 '살인검거'로 변경
    '강도.1': '강도검거',   # '강도.1'을 '강도검거'로 변경
    '강간강제추행.1': '강간강제추행검거', # '강간강제추행.1'을 '강간강제추행검거'로 변경
    '절도.1': '절도검거',   # '절도.1'을 '절도검거'로 변경
    '폭력.1': '폭력검거'    # '폭력.1'을 '폭력검거'로 변경
}, inplace=True)

# 변경된 데이터프레임 확인
Crime_Seoul.head()
```

### 3.2 불필요한 행과 열을 삭제하자

```python
#'기간'열 삭제 (columns='기간', axis=1, inplace=True)

Crime_Seoul.drop(columns='기간', axis=1, inplace=True)
Crime_Seoul.head()
```

```python
#0,1행 삭제 (index=[0,1], axis=0, inplace=True)

Crime_Seoul.drop(index=[0, 1], axis=0, inplace = True)
Crime_Seoul.head()
```

```python
#기존 인덱스를 버리고 인덱스 0부터 다시 설정 (drop=True, inplace=True)

Crime_Seoul.reset_index(drop=True, inplace=True)
Crime_Seoul.head()
```

### 4-2 각 범죄별 검거율을 계산하여 검거율 열을 추가하자

```python
# 각 컬럼의 데이터 타입을 확인
# 검거율을 계산하기 위해서는 데이터가 정수형이거나 실수형이여야 함
# 이 정수형이 아니라 문자열 형임을 알수 있다. 문자열 형은 수학연산이 불가
Crime_Seoul.info()
```

```python
#각 열의 문자열 형의 데이터를 정수형으로 변경
Crime_Seoul = Crime_Seoul.astype({'발생합계':'int64', '검거합계':'int64',
                                  '살인':'int64', '살인검거':'int64',
                                  '강도':'int64', '강도검거':'int64',
                                  '강간강제추행':'int64', '강간강제추행검거':'int64',
                                  '절도':'int64', '절도검거':'int64',
                                  '폭력':'int64', '폭력검거':'int64'})
Crime_Seoul.head()
```

```python
# 각 컬럼의 데이터 타입이 정수형으로 변경되었음 확인

Crime_Seoul.info()
```

```python
#검거 수/발생 수 * 100
#예시) Crime_Seoul['검거율'] = Crime_Seoul['검거합계']/ Crime_Seoul['발생합계']  * 100
#예시) Crime_Seoul['살인검거율'] = Crime_Seoul['살인검거']/ Crime_Seoul['살인']  * 100

Crime_Seoul['검거율'] = Crime_Seoul['검거합계']/ Crime_Seoul['발생합계']  * 100
Crime_Seoul['살인검거율'] = Crime_Seoul['살인검거']/ Crime_Seoul['살인']  * 100
Crime_Seoul['강도검거율'] = Crime_Seoul['강도검거']/ Crime_Seoul['강도']  * 100
Crime_Seoul['강간강제추행검거율'] = Crime_Seoul['강간강제추행검거']/ Crime_Seoul['강간강제추행']  * 100
Crime_Seoul['절도검거율'] = Crime_Seoul['절도검거']/ Crime_Seoul['절도']  * 100
Crime_Seoul['폭력검거율'] = Crime_Seoul['폭력검거']/ Crime_Seoul['폭력']  * 100

Crime_Seoul.head()
```

### 4-3. 각 범죄의 검거수를 삭제하자.

```python
#각 범죄의 검거 열은 삭제, 검거수는 검거율로 대체해 사용
#(columns=['살인검거','강도검거','강간강제추행검거','절도검거', '폭력검거'], axis=1, inplace=True)

# 각 범죄의 검거 열을 삭제하고 검거수를 검거율로 대체하여 사용
Crime_Seoul.drop(columns=['살인검거', '강도검거', '강간강제추행검거', '절도검거', '폭력검거'], axis=1, inplace=True)

# 불필요한 첫 번째, 두 번째 행 삭제
Crime_Seoul.drop(index=[0, 1], axis=0, inplace=True)

# 변경된 데이터프레임 확인
Crime_Seoul.head()
```

### 4-4. 범죄의 검거율이 가장 높은/낮은 상위 5개구는 어디일까?

```python
#범죄 검거율이 가장 높은 구, 데이터 '검거율'열을 기준으로 정렬하기

Crime_Seoul.sort_values(by='검거율', ascending=False).head()
```

### 5.2 범죄 검거율이 가장 높은 구는 어디인가?

```python
#막대 그래프 그리기
# '검거율'로 소팅하여 막대 그래프 그리기

import matplotlib.pyplot as plt
import seaborn as sns

# '검거율' 계산 (검거합계 / 발생합계 * 100)
Crime_Seoul['검거율'] = (Crime_Seoul['검거합계'] / Crime_Seoul['발생합계']) * 100

# '검거율' 기준으로 데이터 정렬 (내림차순)
Crime_Seoul_sorted = Crime_Seoul.sort_values(by='검거율', ascending=False)

# 그래프 크기 설정
plt.figure(figsize=(12, 6))

# 막대 그래프 그리기
sns.barplot(x='구별', y='검거율', data=Crime_Seoul_sorted, palette='Blues_r')

# 제목 및 레이블 설정
plt.title('서울 각 구별 범죄 검거율 (내림차순 정렬)', fontsize=14)
plt.xlabel('자치구', fontsize=12)
plt.ylabel('검거율 (%)', fontsize=12)

# X축 레이블 회전
plt.xticks(rotation=45)

# 그래프 표시
plt.show()
```
